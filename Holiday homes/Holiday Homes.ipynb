```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
#url='https://www.holidayfrancedirect.co.uk/search?board=sc&d=France&people=2'
```

```python
holidayhomes=[]
for x in range(1,31):
    url='https://www.holidayfrancedirect.co.uk/search?board=sc&d=France&people=2&page='
    r=requests.get(url+ str(x))
    soup=BeautifulSoup(r.content,'html.parser')
    #print(soup.title)
    content=soup.find_all('div',class_='property-grid-item')
    #print(content)
    
    for prop in content:
        name=prop.find('h2').text
        spec=prop.find('p',class_='property-spec').text
        price=prop.find('div',class_='property-pricing').text
        #print(name,spec,price)
        property_info={
            'name':name,
            'spec':spec,
            'price':price

        }
        holidayhomes.append(property_info)
        #print(len(holidayhomes))
    time.sleep(3)    
print(holidayhomes) 
    
```

```python
df=pd.DataFrame(holidayhomes)
df.head()
```
